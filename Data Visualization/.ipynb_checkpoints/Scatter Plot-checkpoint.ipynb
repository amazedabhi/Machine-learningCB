```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
prices = np.array([1,2,3,4])**2
print(prices)
```

```python
#plot of only one axis provided to plt plot
#in this case the indices of the numpy array are taken as x and values at those indices are taken as y-axes
plt.plot(prices,marker='*')
plt.show()
```

## Scatter Plot

```python
x=np.arange(10)
y1=x**2
y2=2*x+3
print(x)
print(y1)
print(y2)
```

```python
plt.scatter(x,y1)
plt.scatter(x,y2)
plt.show()
```

```python
#same properties for line plot can be applied for scatter plot
#adjust the size of the plot
plt.figure(figsize=(5,5))
# scatter plot graph using two features
plt.scatter(x,y1,color='red',label='Apple',marker='o')
plt.scatter(x,y2,color='green',label='Kiwi',linestyle='dashed',marker='*')
#give title to x-axis
plt.xlabel('Time')
#give title to y-axis
plt.ylabel('Price')
#show title of the graph
plt.title('Scattter Plot - Fruits over time')
#get a legend on the graph
plt.legend()
#shows the plotted graph
plt.show()
```
