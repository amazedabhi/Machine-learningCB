```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#Line PLot
x=np.arange(10)
y1=x**2
y2=2*x+3
print(x)
print(y1)
print(y2)
```

```python
#plotting on a graph

plt.plot(x,y1)
plt.show()
```

```python
#available themes for printing graphs
theme = plt.style.available
print(theme)
```

```python
#using style seaborn for printing graph
plt.style.use("seaborn")
```

```python
#plot graph using two features
plt.plot(x,y1,color='red',label='Apple',marker='o')
plt.plot(x,y2,color='green',label='Kiwi',linestyle='dashed',marker='*')
#give title to x-axis
plt.xlabel('Time')
#give title to y-axis
plt.ylabel('Price')
#show title of the graph
plt.title('prices of Fruits over time')
#get a legend on the graph
plt.legend()
#shows the plotted graph
plt.show()
```

```python
#plot graph using two features separately
plt.plot(x,y1,color='red')
plt.show()

plt.plot(x,y2,color='green')
plt.show()
```
